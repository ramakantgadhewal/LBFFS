```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def plotUave(df,umax,name,cid):
    cmap = plt.get_cmap('tab10')
    y = df['Points:1']
    uave = df['u_average']/umax+cid
    plt.plot(uave,y,color=cmap(0),label=name,lw=3)

def plotUexp(df,name,cid):
    cmap = plt.get_cmap('tab10')
    y = df['y/H'].astype(float)
    uave = df['U/Uc'].astype(float)+cid
    plt.scatter(uave,y,color=cmap(1),label=name)    

plt.figure(figsize=(24,8))
plt.rcParams["font.size"] = 28

df = pd.read_csv('UMean.dat',sep='\s+')
df00 = df[df['x/H'] == '2.44E-02']
df01 = df[df['x/H'] == '1.00E+00']
df02 = df[df['x/H'] == '2.02E+00']
df03 = df[df['x/H'] == '3.00E+00']
df04 = df[df['x/H'] == '4.02E+00']
df05 = df[df['x/H'] == '5.00E+00']
df06 = df[df['x/H'] == '6.02E+00']
df07 = df[df['x/H'] == '7.00E+00']
df08 = df[df['x/H'] == '8.02E+00']
df09 = df[df['x/H'] == '9.00E+00']
df10 = df[df['x/H'] == '1.00E+01']

plotUexp(df00,'',0)
plotUexp(df01,'',1)
plotUexp(df02,'',2)
plotUexp(df03,'',3)
plotUexp(df04,'',4)
plotUexp(df05,'',5)
plotUexp(df06,'',6)
plotUexp(df07,'',7)
plotUexp(df08,'',8)
plotUexp(df09,'',9)
plotUexp(df10,'',10)


df = pd.read_csv('x00.csv')
umax = df['u_average'].max()
plotUave(df,umax,'x=0',0)

df = pd.read_csv('x01.csv')
plotUave(df,umax,'x=1',1)

df = pd.read_csv('x02.csv')
plotUave(df,umax,'x=2',2)

df = pd.read_csv('x03.csv')
plotUave(df,umax,'x=3',3)

df = pd.read_csv('x04.csv')
plotUave(df,umax,'x=4',4)

df = pd.read_csv('x05.csv')
plotUave(df,umax,'x=5',5)

df = pd.read_csv('x06.csv')
plotUave(df,umax,'x=6',6)

df = pd.read_csv('x07.csv')
plotUave(df,umax,'x=7',7)

df = pd.read_csv('x08.csv')
plotUave(df,umax,'x=8',8)

df = pd.read_csv('x09.csv')
plotUave(df,umax,'x=9',9)

df = pd.read_csv('x10.csv')
plotUave(df,umax,'x=10',10)



#plt.legend(bbox_to_anchor=(1.05,1),loc='upper left', borderaxespad=0.5, fontsize=18)
plt.ylim(0,3)
plt.xlim(-1,11)
plt.xlabel('$x/H$ (-), $u/u_0$ (-)')
plt.ylabel('$y/H$ (-)')
plt.xticks(np.arange(0,11))


plt.show()
```
