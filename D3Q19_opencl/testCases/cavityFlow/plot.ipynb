```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(6,6))
plt.rcParams["font.size"] = 16
plt.xlabel('$u$ (m/s)')
plt.ylabel('$y$ (m)')


df = pd.read_csv('u_x05.csv')

xs=df['Points:1'].min()
xl=df['Points:1'].max()
dx = (xl-xs)/(len(df['Points:1'])-1)
xl = 0.5*dx+xl
xs = xs-0.5*dx
L = xl-xs

plt.plot(df['u'],df['Points:1']/L,label='LBM')


df = pd.read_csv('Ghia.dat')
plt.plot(df['u'],df['y'],label='Ghia', marker='o')

plt.legend()
plt.show()


```
