```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(6,6))
plt.rcParams["font.size"] = 16
plt.xlabel('$u$ (m/s)')
plt.ylabel('$y$ (m)')


df = pd.read_csv('u.csv')
plt.plot(df['u'],df['Points:1'],label='LBM',marker='o')

xs = df['Points:1'].min()
xl = df['Points:1'].max()
dx = (xl-xs)/(len(df['Points:1'])-1)

xl = 0.5*dx+xl
xs = xs-0.5*dx
h = xl-xs

nu = 0.01
dpdx = 0.08
uave = 0.5*dpdx/nu

y = np.linspace(0, h, 100)
u = uave*y*(h -y)
plt.plot(u,y,label='Analytical')

plt.legend()
plt.show()


```
