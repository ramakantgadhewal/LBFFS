```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def plotCx(df,cid=0,name=""):
    df.columns=['t', 'y']

    dfPart = df[df['t']>5]
    fit = np.polyfit(dfPart['t'],dfPart["y"],0)
    ave = np.poly1d(fit)(dfPart['t'])

    cmap = plt.get_cmap('tab10')

    plt.plot(df['t'],df['y'],label=name+'Ave: %.2f'%(ave[0]),color=cmap(cid),linestyle='-')
    plt.plot(dfPart['t'],ave,color=cmap(cid),linestyle='--')

def plotCy(df,cid=0,name=""):
    df.columns=['t', 'y']

    cmap = plt.get_cmap('tab10')

    plt.plot(df['t'],df['y'],label=name,color=cmap(cid),linestyle='-')


df = pd.read_table('t_Cx.dat')
plotCx(df)

plt.xlabel('Time (s)')
plt.ylabel('$C_x$ (-)')
plt.legend()
plt.show()

# ---------------
df = pd.read_table('t_Cy.dat')
plotCy(df)

plt.xlabel('Time (s)')
plt.ylabel('$C_y$ (-)')
plt.legend()
plt.show()



```
